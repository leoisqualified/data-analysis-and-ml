```python
#import the required packages
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
```

```python
#loading the dataset
df = pd.read_csv('/content/drive/MyDrive/BrentOilPrices.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
#splitting the date column
df[['Day','Month','Year']] = df['Date'].str.split('-',expand=True)
df.head()
```

```python
df = pd.get_dummies(df,columns=['Month'])
df.head()
```

```python
#checking if all the months have been encoded
df.columns
```

```python
df.drop(columns='Date', inplace=True)
df.head()
```

```python
#dictionary mapping the encoded month columns with the original columns
month_dict = {
    'Month_Jan':'January',
    'Month_Feb':'February',
    'Month_Mar':'March',
    'Month_Apr':'April',
    'Month_May':'May',
    'Month_Jun':'June',
    'Month_Jul':'July',
    'Month_Aug':'August',
    'Month_Sep':'September',
    'Month_Oct':'October',
    'Month_Nov':'November',
    'Month_Dec':'December'
}
```

```python
#renaming month columns
df = df.rename(columns=month_dict)
df.head()
```
