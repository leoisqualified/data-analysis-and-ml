```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('/content/Electric_Vehicle_Population_Data.csv')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows',None)
```

```python
df.columns
```

**Checking for null values and filling with zero and creating new columns**

```python
df['Fullname'] = df['Make'] + '-' + df['Model']
df['Location'] = df['City'] + '-' + df['State'] + ',' + df['County']
df.isnull().sum()
```

```python
df1 = df.fillna(value=0)
```

```python
df1.head(50)
```

```python
df1[df1['Fullname']=='KIA-SOUL']
```

# Vehicles and their Electric Range

```python
unique_name = df1['Fullname'].unique()
sorted_electric_range = df1.groupby('Fullname')['Electric Range'].max().sort_values(ascending = False)
car_vs_electricrange = pd.DataFrame({'Car':unique_name})
car_vs_electricrange['Electric Range'] = sorted_electric_range.values

car_vs_electricrange = car_vs_electricrange[car_vs_electricrange['Electric Range'] != 0]
car_vs_electricrange
```

```python
# @title Electric Range

from matplotlib import pyplot as plt
car_vs_electricrange['Electric Range'].plot(kind='hist', bins=20, title='Electric Range')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

# **Electric Utility And Location**

```python
electric_utilities = df1['Electric Utility'].unique()
utility_list = []
location_list = []

for utility in electric_utilities:
  locations_for_utility = df1[df1['Electric Utility']== utility]['Location'].unique()
  utility_list.append(utility)
  location_list.append(locations_for_utility)

utility_vs_location = pd.DataFrame({'Utility': utility_list, 'Location': location_list})
utility_vs_location = utility_vs_location[utility_vs_location['Utility']!=0]
utility_vs_location.reset_index(drop=True, inplace=True)
utility_vs_location
```

# CountPlot For Each Vehicle

```python
plt.figure(figsize=(10,20))
sns.countplot(y='Fullname',legend=False, data=df1)
```

```python
plt.figure(figsize = (10,8))
sns.barplot(x=utility_vs_location['Location'].apply(len),y=utility_vs_location['Utility'])
sns.set_style('darkgrid')
plt.xlabel('Number of Locations')
plt.ylabel('Electric Utility')
plt.show()
```
