```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('/content/Electric_Vehicle_Population_Data.csv')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows',None)
df.head(100)
```

```python
df.columns
```

**Checking for null values and filling with zero**

```python
df['Fullname'] = df['Make'] + '-' + df['Model']
df.isnull().sum()
```

```python
df1 = df.fillna(value=0)
df1[df1['Fullname']=='HYUNDAI-KONA']
```

```python
df2 = df1.sort_values(by=['Electric Range'], ascending = False)
df2.head()
```

```python
df2.describe()
```

```python
df2.shape
```

# **Data Visualization**

```python
plt.figure(figsize=(10,20))
plt.title('Countplot For Cars', fontsize=15)
sns.countplot(y='Fullname',data=df2)
```

```python
plt.figure(figsize = (10,15))
sns.barplot(y='Fullname',x='Electric Range',data=df1)
```
